```python
file = datasets.iloc[0]['file_url']
print(file)
data = utl.get_csv_data_from_url(file)
data.head()
```

```python
# Where should I live? 

#get columns we want using tuple-column indexing 
r = [data[x] for x in [           ('Geographic Area'),
                                  ('Females','Age 25 to 29','Christian'),
                                  ('Males','Age 25 to 29','Christian')
                                 ]]
r1 = pandas.concat(r,axis = 1)
#coerse types 
r1[('Females','Age 25 to 29','Christian')] = r1[('Females','Age 25 to 29','Christian')].map(lambda x: int(x))
r1[('Males','Age 25 to 29','Christian')] = r1[('Males','Age 25 to 29','Christian')].map(lambda x: int(x))

# filter for large enough places 
r1= r1[r1[('Females','Age 25 to 29','Christian')] > 300]

# calculate the ratio 
r1['ratio'] = r1[('Females','Age 25 to 29','Christian')] /r1[('Males','Age 25 to 29','Christian')]

r1.sort_values('ratio',ascending = False).head()
```
